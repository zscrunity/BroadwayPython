```python
bill = ""
all_total = 0
n = int(input("Enter n= "))
for i in range(n):
    name = input("Enter name = ")
    price = int(input("Enter price = "))
    qty = int(input('Enter quantity = '))
    total = price * qty
    bill = bill + name +" " + str(price) +" " + str(qty) + " " + str(total) + "\n"
    all_total = all_total + total

print(bill)
```

```python
# WAP to detect if the given no is prime or composite
#for else loop
# prime = 2,3,5,7,11,13
# composite = 4,6,8,9,10

n = int(input("Enter n = "))
i = 2
while i < n:
    if n % i == 0:
        print("The no is Composite no ")
        break
    i = i+1
        
else:
    print("The no is prime")
```
